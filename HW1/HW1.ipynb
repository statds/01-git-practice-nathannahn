```python
import random

def montyHallTrial(ndoors, switch):

    #generate doors given ndoors
    doorList = list(range(1, ndoors+1)) 

    #generate door that contains car
    winDoor = random.choice(doorList)
    #print("Winning Door: Door " + str(winDoor))

    #generate door contestent will choose
    chooseDoor = random.choice(doorList) 
    #print("Chosen Door: Door " + str(chooseDoor))

    #Generates a list of doors that have a goat and have not been picked
    revealDoorList = list()
    for x in doorList:
        if x == winDoor or x == chooseDoor:
            
            continue
        else:
            revealDoorList.append(x)

    #chooses door from that list to reveal to contestant
    revealDoor = random.choice(revealDoorList) 
    #print("Revealed Goat Door: Door " + str(revealDoor))

    #don't switch doors
    if switch == False:

        #if the winning door is the same as the chosen door
        if winDoor == chooseDoor:
            return True
            
        else:
            return False


    #switch doors     
    else:

        #generate a list of values that contains doors that are not the revealed goat door or the door picked by particpant
        finalDoorList = list()
        for x in doorList:
            if x == revealDoor or x == chooseDoor:
                continue
            else:
                finalDoorList.append(x)
                
        #generates a random value from the remaining doors 
        newChoiceDoor = random.choice(finalDoorList)
        
        #if the switched door is the winning door, return true
        if newChoiceDoor == winDoor:
            return True
        else:
            return False

#functions counts how many times the montyHallTrial returns true 
def montyHallTrials(ndoors, trials):
    counterSwitch = 0
    counterNoSwitch = 0
    for x in range(1, trials):
        if montyHallTrial(ndoors, True) == True:
            counterSwitch = counterSwitch + 1
        else:
            counterSwitch = counterSwitch
    
    for x in range(1, trials):   
        if montyHallTrial(ndoors, False) == True:
            counterNoSwitch = counterNoSwitch + 1
        else:
            counterNoSwitch = counterNoSwitch
    
    #return first the proportion of wins for not switch and then the proportion fo wins for switching
    return counterNoSwitch/1000, counterSwitch/1000
    
    
                
#print(montyHallTrial(3, True))
#print(montyHallTrial(5, True))

print("Proportion of wins for 3 doors when not switching doors: " + str(montyHallTrials(3,1000)[0]))
print("Proportion of wins for 3 doors when switching doors: " + str(montyHallTrials(3,1000)[1])) 

print("\nProportion of wins for 5 doors when switching doors: " + str(montyHallTrials(5,1000)[0])) 
print("Proportion of wins for 5 doors trials when switching doors: " + str(montyHallTrials(5,1000)[1])) 
```

# Question 4:

 Write a function to do a Monte Carlo approximation of pi. The function takes a Monte Carlo sample size n as input, and returns a point estimate of pi and a 95% confidence interval. Apply your function with sample size 1000, 2000, 4000, and 8000. Comment on the results.

```python
import random

def monteCarlo(n):

    #create variables for number of points 
    circlePoints = 0
    squarePoints = 0

    for x in range(n):

        #randomly generates x and y values from (-1,1) which are the bounds.
        randX = random.uniform(-1, 1)
        randY = random.uniform(-1, 1)

        #calculates net distance from the origin
        distOrigin = randX**2 + randY**2


        #check to see if randX and randY lie within circle
        if distOrigin <= 1:
            circlePoints = circlePoints + 1
        

        #number of points generated in square (everything is generated in there)
        squarePoints = squarePoints + 1

    
    #formula for estimate of pi
    pi = 4 * (circlePoints/squarePoints)

    #formula for calculating standard deviation of the estimate of pi
    error = 1.96 * ((pi*(1-(circlePoints/squarePoints))/n)/2)

    #return pi and confidence interval
    return pi, (pi-error, pi+error)



print("The Monte Carlo approximation of pi for a sample size of 1000 is: " + str(monteCarlo(1000)[0]))
print("The corresponding confidence interval for that estimate is:  " + str(monteCarlo(1000)[1]))
print("\nThe Monte Carlo approximation of pi for a sample size of 2000 is: " + str(monteCarlo(2000)[0]))
print("The corresponding confidence interval for that estimate is:  " + str(monteCarlo(2000)[1]))
print("\nThe Monte Carlo approximation of pi for a sample size of 4000 is: " + str(monteCarlo(4000)[0]))
print("The corresponding confidence interval for that estimate is:  " + str(monteCarlo(4000)[1]))
print("\nThe Monte Carlo approximation of pi for a sample size of 8000 is: " + str(monteCarlo(800)[0]))
print("The corresponding confidence interval for that estimate is:  " + str(monteCarlo(8000)[1]))
```
