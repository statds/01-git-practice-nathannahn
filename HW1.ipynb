````python
# Question 1

Write a function to demonstrate the Monty Hall problem through simulation. The function takes two arguments #ndoors and #ntrials, representing the number of doors in the experiment and the number of trails in a simulation, respectively. The function should return the proportion of wins for both the switch and no-switch strategy. Apply your function with 3 doors and 5 doors, both with 1000 trials. Include sufficient text around the code to explain your them.

```{python}

import random

def montyHallTrial(ndoors, switch):

    #generate doors given ndoors
    doorList = list(range(1, ndoors+1)) 

    #generate door that contains car
    winDoor = random.choice(doorList)
    #print("Winning Door: Door " + str(winDoor))

    #generate door contestent will choose
    chooseDoor = random.choice(doorList) 
    #print("Chosen Door: Door " + str(chooseDoor))

    #Generates a list of doors that have a goat and have not been picked
    revealDoorList = list()
    for x in doorList:
        if x == winDoor or x == chooseDoor:
            
            continue
        else:
            revealDoorList.append(x)


    #chooses door from that list to reveal to contestant
    revealDoor = random.choice(revealDoorList) 
    #print("Revealed Goat Door: Door " + str(revealDoor))

    #don't switch doors
    if switch == False:

        #if the winning door is the same as the chosen door
        if winDoor == chooseDoor:
            return True
            
        else:
            return False


    #switch doors     
    else:

        #generate a list of values that contains doors that are not the revealed goat door or the door picked by particpant
        finalDoorList = list()
        for x in doorList:
            if x == revealDoor or x == chooseDoor:
                continue
            else:
                finalDoorList.append(x)
                
        #generates a random value from the remaining doors 
        newChoiceDoor = random.choice(finalDoorList)
        
        #if the switched door is the winning door, return true
        if newChoiceDoor == winDoor:
            return True
        else:
            return False

#functions counts how many times the montyHallTrial returns true 
def montyHallTrials(ndoors, trials):
    counterSwitch = 0
    counterNoSwitch = 0
    for x in range(1, trials):
        if montyHallTrial(ndoors, True) == True:
            counterSwitch = counterSwitch + 1
        else:
            counterSwitch = counterSwitch
    
    for x in range(1, trials):   
        if montyHallTrial(ndoors, False) == True:
            counterNoSwitch = counterNoSwitch + 1
        else:
            counterNoSwitch = counterNoSwitch
    
    print("Success Rate No Switch: " + str(counterNoSwitch/1000)) 
    print("Success Rate Switch: " + str(counterSwitch/1000)) 
    
                
#print(montyHallTrial(3, True))
#print(montyHallTrial(5, True))
montyHallTrials(3, 1000)
montyHallTrials(5, 1000)
````
